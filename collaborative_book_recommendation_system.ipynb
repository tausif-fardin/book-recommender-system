```python
from keras.layers import Input, Dense
from keras.models import Model
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
# Encoder
input_book = Input(shape=(X_train.shape[1],))
encoded = Dense(128, activation='relu')(input_book)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dense(32, activation='relu')(encoded)
```

```python
# Decoder
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(X_train.shape[1], activation='sigmoid')(decoded)
```

```python
# Autoencoder Model
autoencoder = Model(input_book, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping

# Training the Autoencoder
autoencoder.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, X_test), callbacks=[EarlyStopping(patience=10)])
```

```python
# Evaluate the model
autoencoder.evaluate(X_test, X_test)
```
